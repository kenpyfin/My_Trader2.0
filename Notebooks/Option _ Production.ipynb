```python

table = pd.DataFrame()
my_price = get_price_data([symbol],method = method,robinhood = robinhood).iloc[-1].Close
for i in range(0,11,1):
    try:
        if pc == "call":
            i = -i
#     strike_list.append(my_price*(1-i/100.0))
        op = option_trading(robinhood,symbol,pc,bs,exp_pos,strike = my_price*(1-i/100.0))
        op_table = op.current_Option()
        op_table_tick = op.get_option()["min_ticks"].iloc[0]
    #     op_table.set_value("Above_tick",op_table_tick["above_tick"])
        op_table.set_value("Below_tick",op_table_tick["below_tick"])
    #     op_table.set_value("Cutoff_price",op_table_tick["cutoff_price"])
        op_table.set_value("Experition",op.get_exp_date()[exp_pos])
        op_table.set_value("Strike",op.strike)
        op_table.set_value("Symbol",symbol)
        table = table.append(op_table,ignore_index=True)
    #     time.sleep(2)
    except Exception as e:
        print ("Error occor: %s on %s"%(e,i/100.0))
        pass
```
